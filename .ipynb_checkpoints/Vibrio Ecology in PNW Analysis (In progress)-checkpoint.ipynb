This dataset was taken from NOAA at https://www.fisheries.noaa.gov/inport/item/17958.
The purpose of the study was to invetigate how envirnmental factors affect vibrio populations in the pacific northwest due to the increase in cases of vibrio-related gastroenteritis ue to consumption of raw oysters. TDH+ stands for '
Thermostable direct hemolysin (TDH), a major virulence factor of Vibrio parahaemolyticus'. This toxin beomes deactivated at around 60 to 70C, but becomes reactivated if the temperature crosses 80C. The purpose of this analysis is to investigate how NOAA collected data can explain the increase in TDH positive and overall vibrio populations in the PNW area.

I want to predict the amount of TDH+ based off of:
Salinity
Chlorophyll
Oyster Tissue Temperature
Air Temperature
Surface Water Temperature
Water Temperature At -3 Feet
sklearn's regression random forest will be used for this prediction

```python
#Define target variable & features
features=['Salinity','Chlorophyll','Oyster Tissue Temperature','Air Temperature','Surface Water Temperature','Water Temperature At -3 Feet']
features_df=pd.DataFrame()
for name in features:
    features_df[name]=df[name]
```

```python
features_df
```

```python
X=features_df
y=df['Tdh+ Vibrio Parahaemolyticus, Water']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_RF_regression=RandomForestRegressor(n_estimators=5)
model_RF_regression.fit(X_train, y_train)
```

```python
y_predictions=model_RF_regression.predict(X_test)
```

```python
y_predictions
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predictions)
```
