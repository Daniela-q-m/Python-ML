This dataset was taken from NOAA at https://www.fisheries.noaa.gov/inport/item/17958.
The purpose of the study was to invetigate how envirnmental factors affect vibrio populations in the pacific northwest due to the increase in cases of vibrio-related gastroenteritis ue to consumption of raw oysters. TDH+ stands for '
Thermostable direct hemolysin (TDH), a major virulence factor of Vibrio parahaemolyticus'. This toxin beomes deactivated at around 60 to 70C, but becomes reactivated if the temperature crosses 80C. The purpose of this analysis is to investigate how NOAA collected data can explain the increase in TDH positive and overall vibrio populations in the PNW area.

```python

sns.pairplot(df, vars=['Tdh+ Vibrio Parahaemolyticus, Water','Tdh+ Vibrio Parahaemolyticus, Net Tow','Tdh+ Vibrio Parahaemolyticus, Oyster','Surface Water Temperature','Water Temperature At -3 Feet','Salinity','Chlorophyll'])
```

```python
These plots above show:
    -Linear relationship between surface water temperature and temperature at -3 feet
    -Normal-like distribution of  water temperature vs chlorophyll, vibrio counts and water temperature (temperatures around 15C showed highest vibrio populations)
    -A right skewed distribution of salinity vs chlorophyll (higher salinity, higher chlorophyll)
    -A right skewed distribution showing higher tdh vibrio at higher salinity
    -A left skewed distribution showing lower chlorophyll values associated with higher vibrio populations
    
```
